# Here we will load in a CSV file as a Data Frame


```python
import pandas as pd

iris_data = pd.read_csv("/content/drive/MyDrive/PGSS2021/L1-762021/iris.csv", header=None )
iris_feature_names = ["Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width", "Species"]

iris_data.columns = iris_feature_names
iris_data
```

```python
iris_data.iloc[4, :]
```

```python
iris_data.iloc[3, 2]
```

```python
iris_data.iloc[:,0]
```

```python
iris_data["Sepal.Length"] = iris_data["Sepal.Length"].astype("float32")
iris_data["Sepal.Width"] = iris_data["Sepal.Width"].astype("float32")
iris_data["Petal.Length"] = iris_data["Petal.Length"].astype("float32")
iris_data["Petal.Width"] = iris_data["Petal.Width"].astype("float32")

iris_data["Species"] = iris_data["Species"].astype("category")
iris_data.info()
```

# Let's make some summary statistics and report on the same 

```python
iris_data["Species"].value_counts()
```

```python
iris_data.groupby('Species').agg({'Sepal.Length': ['min', 'max', 'mean'], 'Sepal.Width': ['min', 'max', 'mean'], 
                                  'Petal.Length': ['min', 'max'], 'Petal.Width': ['min', 'max']})
```

```python
print("Min values by column \n" + str(iris_data['Petal.Length'].aggregate( 'min' )) + '\n\n')
print("Max values by column \n" + str(iris_data['Petal.Width'].aggregate( 'max' )))
```
